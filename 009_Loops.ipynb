## Python Loops

    - loop statement allows us to execute a statement or group of statements multiple time.
    - python provides following type of loops:
        -- while loop
        -- for loop
        -- nested loop

### While Loop

    while expression:
        statement(s)
        
    - while loop statement executes as long as a given condition is true.
    - statement(s) may be a single statement or a block of statements.
    - condition may be any expression.

```python
num=0
while num<12:
    print(num)
    num+=1
```

### Else statement with While Loop

    - works like if-else statement.
    - else statement is executed when the condition becomes false.

```python
cnum=0
while cnum<7:
    print(cnum)
    cnum+=1
else:
    print("You Reached the Maximum limit")
```

### For Loop

    for iter_var in sequence:
        statements(s)
        
    - for statement iterate over the item of sequences (list, tuple etc).

```python
for i in list(range(0,12,2)):
    print(i)
```

```python
ls1=["India","Brazil","Canada","Germany","Mexico"]
for i in ls1:
    print(i)
```

```python
for i in "Banana":
    print(i)
```

### Iterating by Sequence Index

```python
for idx in range(len(ls1)):
    print(ls1[idx])
```

### Else Statement with For Loop

    - else statement is executed when the condition becomes false.

```python
ls2=[12,33,17,18,13,9,5,24]
for val in ls2:
    if val%7==0:
        print(f"{val} is Divisible by 7.")
        break
else:
    print("Not a Single Number which is Divisble by 7.")
    
```

```python
ls3=[12,33,17,18,14,9,5,24]
for val in ls3:
    if val%7==0:
        print(f"{val} is Divisible by 7.")
        break
else:
    print("Not a Single Number which is Divisble by 7.")
```

### Nested Loops

    - we can use one loop inside another loop.

```python
for i in list(range(1,6)):
    for j in list(range(1,11)):
        print(i*j,end=' ')
    print()
```

```python
n=6
r=1
while r<n:
    c=1
    while c<=r:
        print(c,end=' ')
        c+=1
    print()
    r+=1
        
```

### Loop Control Statement

    - Break
    - Continue
    - Pass

#### Break Statement

    - break statement can be used in both while and for loops.
    - used for premature termination of the current loop.
    - in case of nested loop, break statement stops the execution of the innermost loop.

```python
for i in range(10):
    if i==4:
        break
    print(i)
```

```python
wcnt=0
while wcnt<12:
    print(wcnt)
    wcnt+=1
    if wcnt==7:
        break
```

```python
for i in range(1,6):
    for j in range(1,12):
        val=i*j
        if val>=16:
            break
        print(val,end=' ')
    print()
```

#### Continue Statement

    - returns the control to the beginning of the current loop.
    - Continue Statement can be used in both while and for loops.

```python
for i in range(1,11):
    if i%2==0:
        continue
    print(i)
```

```python
nn=1
while nn<=10:
    nn+=1
    if nn%2!=0:
        continue
    print(nn)
```

#### Pass Statement

    - pass statement is a null operation
    - used as a placeholder for future code.
    - used with loops, function definitions, class definitions, or in if statements.

```python
for i in ['This', 'is','Pass', 'Statement']:
    if i=='Pass':
        pass
        print(i)
```

```python
class Person:
    pass
```

```python
def toDo():
    pass
```

### Infinite Loop

    - A loop becomes infinite loop if a given condition never become FALSE.
    - useful in client/server programming.

```python
count=0
while True:
    print(f"Infinite {count}") #interrupt the kernel or it will be keep running until an external factor interferes in the execution flow. 
    count+=1
```
